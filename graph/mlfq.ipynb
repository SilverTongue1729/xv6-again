# MLFQ Scheduling Analysis

## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Plot

```python
# Initialize an array of 10 lists to store queue information for each process
nproc = 10
process_queues = [[] for _ in range(nproc)]

# Read the lines from the mlfq_data.txt file
with open('mlfq_data.txt', 'r') as file:
    lines = file.readlines()

# Parse each line and update the corresponding process queue
for tick, line in enumerate(lines):
    parts = line.strip().split(':')
    time_tick = int(parts[0])
    parts = parts[1].split(',')
    for part in parts:
        pid, queue = map(int, part.split())
        process_queues[pid - 1].append((tick, queue))

# Plot the graph
colors = ['b'] * 5 + ['g', 'r', 'c', 'm', 'y']

plt.figure(figsize=(12, 6))
for pid, queue_data in enumerate(process_queues):
    time_ticks, queues = zip(*queue_data)
    plt.plot(time_ticks, queues, label=f'Process {pid + 1}', color=colors[pid])
    
    if time_ticks:
        plt.scatter(time_ticks[-1], queues[-1], color=colors[pid], marker='o')


plt.yticks([0, 1, 2, 3])

plt.xlabel('Number of Ticks')
plt.ylabel('Queue Number')
plt.legend()
plt.title('MLFQ Visualization')
plt.grid(True)
plt.show()
```
