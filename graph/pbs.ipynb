# PBS Analysis

## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Plot

```python
# Initialize an array to store priority information for each process
nproc = 10
process_priorities = [[] for _ in range(nproc)]

# Read the lines from the pbs_data.txt file
with open('pbs_data.txt', 'r') as file:
    lines = file.readlines()

# Parse each line and update the corresponding process priority list
for tick, line in enumerate(lines):
    parts = line.strip().split(':')
    time_tick = int(parts[0])
    parts = parts[1].split(',')
    for part in parts:
        pid, priority = map(int, part.split())
        process_priorities[pid - 1].append((tick, priority))

# Plot the graph
# colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'orange', 'pink', 'brown']
colors = ['b'] * 5 + ['g', 'r', 'c', 'm', 'y']

plt.figure(figsize=(12, 6))
for pid, priority_data in enumerate(process_priorities):
    time_ticks, priorities = zip(*priority_data)
    plt.plot(time_ticks, priorities, label=f'Process {pid + 1}', color=colors[pid])
    
    if time_ticks:
        plt.scatter(time_ticks[-1], priorities[-1], color=colors[pid], marker='o')

plt.yticks(range(0, 101, 10))  # Assuming you want to show every 10th priority level

plt.xlabel('Ticks')
plt.ylabel('Priority')
plt.legend()
plt.title('Priority Based Scheduling Visualization')
plt.grid(True)
plt.show()
```
